```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

a=sns.load_dataset("iris")
a.columns
```

```python
sns.relplot(x='sepal_length',y='sepal_width',data=a,hue='species')
```

```python
sns.catplot(x='sepal_length',y='sepal_width',data=a)
```

```python
sns.displot(a)
```

```python
a=sns.FacetGrid(a,col="species")
a.map(plt.hist,"sepal_length")
```

```python
sns.set(style="darkgrid")
a=sns.load_dataset("iris")
a=sns.PairGrid(a)
a.map(plt.scatter)
```

```python

a=sns.load_dataset('iris')
sns.boxplot(x='sepal_length',y='sepal_width',data=a)
```

```python
c=sns.color_palette()
sns.palplot(c)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=a,kind='reg')
```

```python
sns.kdeplot(x='sepal_length',y='sepal_width',data=a)
```

```python
sns.pairplot(a)
```

```python
sns.rugplot(x='sepal_length',y='sepal_width',data=a)
```

```python
sns.rugplot(a['petal_length'])
```

```python
sns.barplot(x='sepal_length',y='sepal_width',data=a,estimator=np.median)
```

```python
sns.countplot(x='sepal_length',data=a)
```

```python
sns.violinplot(x='sepal_length',y='sepal_width',data=a,split=True)
```

```python
sns.stripplot(x='sepal_length',y='sepal_width',data=a,jitter=True,dodge=True)
```

```python
sns.violinplot(x='sepal_length',y='sepal_width',data=a)
sns.swarmplot(x='sepal_length',y='sepal_width',data=a,color='red')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

a=sns.load_dataset("diamonds")
a.columns
```

```python
sns.relplot(x='carat',y='price',data=a)
```

```python
sns.catplot(x='carat',y='price',data=a)
```

```python
sns.displot(a['carat'])
```

```python
a=sns.load_dataset("diamonds")
a=sns.FacetGrid(a,col='color')
a.map(plt.hist,"carat")
```

```python
a=sns.load_dataset("diamonds")
a=sns.PairGrid(a)
a.map(plt.scatter)
```

```python
sns.set(style="white",color_codes=True)
a=sns.load_dataset("diamonds")
sns.boxplot(x="carat",y="price",data=a)
```

```python
sns.jointplot(x="carat",y="price",data=a)
```

```python
sns.kdeplot(a['table'])
```

```python
sns.rugplot(a['depth'])
```

```python
sns.pairplot(a)
```

```python
sns.barplot(x="carat",y="price",data=a,estimator=np.mean)
```

```python
sns.countplot(x='carat',data=a)
```

```python
sns.violinplot(x="carat",y="price",data=a,split=True)
```

```python
plt.figure(figsize=(8,5))
sns.stripplot(x="carat",y="price",data=a,jitter=True,dodge=True)
```
