```python
# MISSING DATA
df=pd.DataFrame({'A':[1,2,np.nan],
                 'B':[5,np.nan,np.nan],
                 'C':[1,2,3]})
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
# thresh takes integer value which tells minimm amount of na values to drop
df.dropna(thresh=2)
```

```python
df.fillna(value='FILL VALUE')
```

```python
df['A'].fillna(value=df['A'].mean())
```

```python
# group by
# create dataframe

data={'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
      'Person':['Sam','charlie','amy','vaneessa','carl','sarah'],
      'Sales':[200,120,340,124,243,350]}
df=pd.DataFrame(data)
df
```

```python
df.groupby('Company')
```

```python
# u can save this object as a new variable:
by_comp=df.groupby("Company")
by_comp
```

```python
by_comp.mean()
```

```python
df.groupby('Company').mean()
```

```python
# more eg of aggregate method
by_comp.std()
```

```python
by_comp.min()
```

```python
by_comp.max()
```

```python
by_comp.count()
```

```python
by_comp.describe()
```

```python
by_comp.describe().transpose()
```

```python
by_comp.describe().transpose()['GOOG']
```
