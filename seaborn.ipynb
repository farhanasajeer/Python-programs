```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# relplot
# it used to creating relational plot
# statistical data
a=sns.load_dataset("flights")
sns.relplot(x="passengers", y="month", data=a)
```

```python
a=sns.load_dataset("flights")
sns.relplot(x="passengers", y="month",hue="year", data=a)
```

```python
b=sns.load_dataset("tips")
a=sns.load_dataset("flights")
sns.relplot(x="time", y="tip", data=b, kind="line")
```

```python
# CAT PLOT CATOGERICAL DATA
# catplot used for creating categorical plot

sns.catplot(x="day",y="total_bill",data=b)
```

```python
sns.catplot(x="day",y="total_bill",kind="violin", data=b)
```

```python
# DISTRIBUTION
from scipy import  stats

# Displot used for creating distribtion plots
# univariant
c=np.random.normal(loc=5, size=100,scale=2)
sns.displot(c)
```

```python
# FACTGRID
a=sns.load_dataset("iris")

# factgrid class helps in visualizing distribution of one variable as well as the realationship bw multiple variable seperately within subset of yur data
b=sns.FacetGrid(a,col="species")
b.map(plt.hist,"sepal_length")
```

```python
# pairgrid
a=sns.load_dataset("flights")
# this class maps each variable in a dataset onto a column and row in agrid of multiple axes
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style="darkgrid")
a=sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style="dark")
a=sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
# BOXPLOT
sns.set(style="white",color_codes=True)
a=sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill",data=a)
```

```python
# remove axis line

sns.set(style="white",color_codes=True)
a=sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill",data=a)
sns.despine(offset=10,trim=True)
```
