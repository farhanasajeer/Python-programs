```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# load abuilt in dataset based on US state car crash  percentages
crash_df = sns.load_dataset('car_crashes')

# distribution plot
# provides a way to look at a univariate distribution . A
# univeriate distribution provides a distribution for one variable
sns.displot(crash_df['not_distracted'], kde=False, bins=25)
```

```python
# join plot
# joinplot compares 2 distributions and plots a scatter plot by default
# as we can see as people tend to speed they also tend to drink & drive
sns.jointplot(x='speeding', y='alcohol',data=crash_df,kind='reg')
```

```python
# KDE PLOT
# a kernal density estimate (KDE) plot is a method for visualizing the distribution of observations in a datset,analagous to ahistogram
# get just the KDE plot

sns.kdeplot(crash_df['alcohol'])
```

```python
# pair plots
# pair plot plots relationships across the entire data frames numerical values
sns.pairplot(crash_df)

# load data on tips
tips_df = sns.load_dataset('tips')
```

```python
# RUG PLOTS

# plots a single column of datapoints in array as sticks on an axis
# with a rug plot you'll see a more dense number of lines where the amount is
# most common. this is like how histograms is taller where values are more common

sns.rugplot(tips_df['tip'])
```

```python
# styling

# you can set styling for your axes and grids
# white,darkgrid,whitegrid,dark,ticks
sns.set_style('white')

# you can use figure sizing from matplotlib
plt.figure(figsize=(8,4))

# change size of labels , lines and other elements to best fit
# how you will present your data (paper,talk,poster)
sns.set_context('paper',font_scale=1.4)

sns.jointplot(x='speeding', y='alcohol',data=crash_df,kind='reg')


# get rid of spines
# you can turn of specific spines with right =true , left=true
# bottom=true, top=true
sns.despine(left=False,bottom=False)
```

```python
# Ctegorical plots
# bar plots

# focus on distributions using categorical data in reference to one of the numerical
# columns

# aggregate categorical data based on a function (mean is default)
# Estimate total bill amount based on sex
# With estimater you can define functions to use other than the mean like those
# provided by Numpy:meadian,std,var,cov or make your own functions

sns.barplot(x='sex',y='total_bill',data=tips_df, estimator=np.median)
```

```python
# COUNT PLOT
# A column plot is like a bar plot, but the estimator is counting
# the no of occurances

sns.countplot(x='sex',data=tips_df)
```

```python
# Boxplot
plt.figure(figsize=(14,9))
sns.set_style('darkgrid')

# A boxplot allows you to compare diffrent variables
# the box shows the quartiles of data the bar in middle is the median and the box extends 1 std deviation from median
# the whiskers extend to all the other data aside from the points that are considered to be outliers
# hue can add another category being sex 
# we see men spend way more on friday versus < women on saturday
sns.boxplot(x='day',y='total_bill',data=tips_df,hue='sex')

# Moves legend to the best position 
plt.legend(loc=0)
```

```python
# violinplot

# it is a combination of boxplot and KDE
# while a box plot corresponds to data points,the violin plot the KDE estimation of the data points
# Split allows you to compare how the catogories compare to each other
sns.violinplot(x='day',y='total_bill',data=tips_df, hue='sex',split=True)
```

```python
# strip plot 
plt.figure(figsize=(8,5))

# the strip plot draws a scatter plot reprasenting all data points wher one variable is catogericacl.
# it is often used to show all observations with a box plot that reprasents the avg distribution
# jitter spreads data points out so that they aren't stacked o top of each other
# Hue breaks data into men and women
# Dodge seperates the men and women data
sns.stripplot(x='day',y='total_bill',data=tips_df, hue='sex',jitter=True,dodge=True) 
```

```python
# swarm plot
# it is like a strip plot, but points are adjusted so they don't overlap
# it looks like a combination of the violin and strip plots
# sns.swarmplot(x='day',y='total_bill',data=tips_df)
# u can stack violin plot with a swarm

sns.violinplot(x='day',y='total_bill',data=tips_df)
sns.swarmplot(x='day',y='total_bill',data=tips_df, color='white')
```

```python
# PALLETTES

plt.figure(figsize=(8,6))
sns.set_style('dark')
sns.set_context('talk')

# u can use matplotlibs color maps for color styling
# https://matplotlib.org/3.3.1/tutorials/colors/colormaps.html
sns.stripplot(x='day',y='total_bill',data=tips_df, hue='sex',palette='seismic')

# add the optionl legend with a location no(best:0,upper right:1,upper left:2,lowerleft:3, lower right:4,
# https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.legend.html)
# or supply a tuple of x & y from lower left

plt.legend(loc=0)
```
